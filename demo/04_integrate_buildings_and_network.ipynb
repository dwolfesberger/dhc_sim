## 1. Load TEASERPLUS project

```python
import os
import pickle
from pathlib import Path
from modelica_builder.model import Model

# Get current working directory
current_dir = os.getcwd()

# Set path to pickle file of TEASERPLUS project class
pickle_file = os.path.join(current_dir, "output", "demo_teaser_pickle.p")

# Load TEASERPLUS project class
prj = pickle.load(open(pickle_file, "rb"))

# get teaser building names
bldg_names = [bldg.name for bldg in prj.buildings]
```

## Edit Modelica network model

- Replace timeseries table with teaser building model
- Set new connections
- Replace _AixLib.Media.Spezialized.Water.ConstantProperties_pT_ with _AixLib.Media.Water_

```python
# set path to model
file_path = os.path.join(current_dir, 'output', 'modelica', 'Heat_network_open_loop_dT_var', 'Heat_network_open_loop_dT_var_inputs.mo')

# Create model class
model = Model(file_path)

def replace_timeseries_with_teaser_load(model, bldg_names):
    """Replace timeseries with teaser model."""
    # Iterate through building names
    for bldg in bldg_names:
        # Get teaser model
        teaser_model = f'teaser.{bldg}.{bldg}_Models.{bldg}_SingleDwelling'
        
        # Get teaser model name
        teaser_model_name = f'{bldg}_teaser'
        
        # Get teaser model QFlowHeat output
        teaser_model_QFlowHeat = f'{teaser_model_name}.QFlowHeat'
        
        # Insert teaser model       
        model.insert_component(teaser_model, teaser_model_name, insert_index=-1)
        
        # Get timeseries connector
        bldg_load_timeseries = f'{bldg}Table.y[1]'
        
        # Get input connector for building substation load
        bldg_network_load_input = f'networkModel.{bldg}Q_flow_input'
        
        # replace timeseries connector with teaser load output
        model.edit_connect(bldg_network_load_input, bldg_load_timeseries, new_port_b=teaser_model_QFlowHeat)
        
        # Get timeseries table
        bldg_load_timerseries_table = f'{bldg}Table'
        
        # Remove timeseries table
        model.remove_component(identifier=bldg_load_timerseries_table)

replace_timeseries_with_teaser_load(model, bldg_names)

model.save_as(file_path)
```
